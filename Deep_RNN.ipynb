```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, LSTM, GRU
```

```python
# We load the given datasets
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)
```

```python
# Pad sequences
x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import numpy as np
model = Sequential([
    Embedding(input_dim=10000, output_dim=32, input_length=100), # Embedding means simply create vectors for each sentences
    SimpleRNN(5, return_sequences=True),   # Layer1 RNN
    SimpleRNN(5),
    Dense(1, activation='sigmoid')
    ])

# Create a sample input
sample_input = np.zeros((1, 100))  # Batch size 1, input length 100

# Explicitly build the model
model.build(input_shape=sample_input.shape)

# Now call summary
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# LSTM Model
model = Sequential([
    Embedding(10000, 32, input_length=100),
    LSTM(5, return_sequences=True),
    LSTM(5),
    Dense(1, activation='sigmoid')
])

model.build(input_shape=(1, 100))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'] )
```

```python
# GRU model
model = Sequential([
    Embedding(10000, 32, input_length=100),
    GRU(5, return_sequences=True),
    GRU(5),
    Dense(1, activation='sigmoid')
])
model.build(input_shape=(1,100))
model.summary()
```

```python
# compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```
