```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, SimpleRNN, LSTM, GRU, Dense
```

```python
# Load the IMDB datasets
num_words = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words = num_words)
```

```python
# Pad Sequences
maxlen = 100
X_train = pad_sequences(x_train, maxlen = maxlen)
X_test = pad_sequences(x_test, maxlen = maxlen)
embedding_dim = 32
```

```python
# Build Bidirectional RNN Model
model = Sequential([
    Embedding(num_words, output_dim=embedding_dim, input_length = maxlen),
    SimpleRNN(5),
    Dense(1, activation='sigmoid')
])

sample_input = np.zeros((1, maxlen))
model.build(input_shape=sample_input.shape)
# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])

model.summary()
```

```python
# Model fit
model.fit(X_train, y_train, epochs=5, batch_size=64, validation_data=(X_test, y_test))
```

```python
# Bidirectional Model
model = Sequential([
    Embedding(num_words, output_dim = embedding_dim, input_length=maxlen),
    Bidirectional(SimpleRNN(5)),
    Dense(1, activation='sigmoid')
])

sample_input = np.zeros((1, maxlen))
model.build(input_shape=sample_input.shape)
# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
#Show model
model.summary()
```
